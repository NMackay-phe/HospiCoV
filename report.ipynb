# HospitCov

Tuto rapide Jupyter pour les noobs: (en gros ya une palette de commandes à gauche, c'est tout ce qu'il y a à savoir en fait)

- ça fonctionne par "cell". Chaque cell peut être type "code" ou "markdown" (ou "raw")
- on navigue dans le notebook avec 2 modes (un peu comme Vim grosso merdo, Dr P va être content) :
   1. un mode pour créer, bouger, copier, coller des cells dans le notebook
   2. un mode où on "rentre" dans la cell pour éditer son contenu


- Les commandes en mode 1:
    - b = créer une cell below
    - a = créer une cell above
    - m = type de la cell en Markdown
    - y = type de la cell en code
    - c/v = copier/coller
    - dd = supprimer la cell


- Touche "enter" pour rentrer dans la cell, après c'est de l'éditeur de texte classique

- Ctrl+Enter pour exécuter une cell
- Echap pour sortir de la cell sans exécuter

Le reste est dans la palette de commande

### Severity

- 80% of laboratory confirmed patients have had mild to moderate disease
- 13.8% have severe disease (dyspnea, respiratory frequency ≥30/minute, blood oxygen saturation≤93%, PaO2/FiO2 ratio <300, and/or lung infiltrates >50% of the lung field within 24-48 hours)
- 6.1% are critical (respiratory failure, septic shock, and/or multiple organ dysfunction/failure).
- Asymptomatic infectionhas been reported, but the majority of the relatively rarecases who are asymptomatic on the date of identification/report wenton todevelop disease.
